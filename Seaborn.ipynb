```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Dataset Load
```

```python
df = sns.load_dataset("tips")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
## Line Chart :- No Line chart use matplotlib plot function 
```

```python
## Histogram
```

```python
sns.histplot(data=df,x="tip",color="tomato",kde=True)
```

```python
sns.histplot(data=df,x="total_bill",color="tomato",kde=True)
```

```python
# Count Plot
```

```python
sns.set_theme()
sns.countplot(data=df,x="day",hue="sex")
plt.title("Customer per Day")
```

```python
## Heatmap
# Used to check the matrix correlation between the numrical values
```

```python
corr = df[["total_bill","tip"]].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

```python
# catplot
# Used Numerical data and Categorical Data
```

```python
sns.catplot(data=df,x="day",y="tip",hue="sex")
```

```python
sns.catplot(data=df,x="day",y="total_bill",hue="sex")
```

```python
# Barplot
```

```python
sns.barplot(data=df,x="day",y="total_bill",hue="sex")
```

```python
average_bill = df.groupby(["day","sex"])["total_bill"].mean().reset_index()
average_bill
```

```python
sns.barplot(data=average_bill,x="day",y="total_bill",hue="sex")
```
